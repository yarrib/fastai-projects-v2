```python
fish_types = 'walleye', 'sauger', 'saugeye','northern pike', 'musky', 'bluegill', 'crappie', 'channel catfish', 'common carp'
path = Path('fish')
```

```python
fish_types
```

```python
import time
if not path.exists():
    path.mkdir()
    for f in fish_types:
        dest = path/f
        dest.mkdir(exist_ok=true)
        time.sleep(1)
        results = search_images_bing(key, f'{f}')
        download_images(dest, urls=results.attrgot('contentUrl'))
```

```python
fns = get_image_files(path)
failed = verify_images(fns)
failed.map(Path.unlink)
```
